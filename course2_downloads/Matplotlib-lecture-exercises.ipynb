```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(4,2))

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

bars = plt.bar(pos, popularity, align='center', alpha=0.7)
bars[0].set_color('red')



plt.xticks(pos, languages, color='grey')
#plt.ylabel('% Popularity', color='grey')
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

#TODO: remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False,left=False,bottom=False,right=False, labelleft=False)
#, labelleft='off'

axes = plt.gca()

for spine in axes.spines.values():
    spine.set_visible(False)


for rect,label in zip(axes.patches, popularity):
    axes.text(rect.get_x() + rect.get_width() /2, rect.get_height() + 5, label, horizontalalignment='center', verticalalignment='bottom')
    
plt.show()
```
